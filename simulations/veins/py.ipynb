```python
import pandas as pd
import matplotlib.pyplot as plt

# Đọc dữ liệu
df = pd.read_csv('/home/ad/Documents/PracticeCpp/global_log.csv')

# Chuyển các cột về kiểu số nếu cần
for col in ['num_fail', 'sum_objective', 'sum_deviation']:
    df[col] = pd.to_numeric(df[col], errors='coerce')
    
print(df)
# Vẽ histogram cho từng cột
plt.figure(figsize=(15, 4))

plt.subplot(1, 3, 1)
plt.bar(df['totalcar'], df['sum_time'], color='skyblue', label='Bar')
plt.plot(df['totalcar'], df['sum_time'], color='blue', marker='o', label='Line')
plt.xlabel('totalcar')
plt.ylabel('sum_time')
plt.title('sum_time vs totalcar')
plt.legend()

plt.subplot(1, 3, 2)
plt.bar(df['totalcar'], df['sum_deviation'], color='orange', label='Bar')
plt.plot(df['totalcar'], df['sum_deviation'], color='red', marker='o', label='Line')
plt.xlabel('totalcar')
plt.ylabel('sum_deviation')
plt.title('sum_deviation vs totalcar')
plt.legend()

plt.subplot(1, 3, 3)
plt.bar(df['totalcar'], df['sum_objective'], color='green', label='Bar')
plt.plot(df['totalcar'], df['sum_objective'], color='darkgreen', marker='o', label='Line')
plt.xlabel('totalcar')
plt.ylabel('sum_objective')
plt.title('sum_objective vs totalcar')
plt.legend()

plt.tight_layout()
plt.show()
```
